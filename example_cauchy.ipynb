```python
from bayes_test import BayesianHypothesisTestCauchy
import pandas as pd
import numpy as np

draws = 2000

study = np.load('group1.npy')

control = np.load('group2.npy')


df1 = pd.DataFrame({'IQ' : control})
df2 = pd.DataFrame({'IQ' : study})
df1['group'] = 'control'
df2['group'] = 'study'

df = pd.concat([df1,df2])

study_test = BayesianHypothesisTestCauchy(df=df,
                                   group1_name='study',
                                   group2_name='control',
                                   value_column='IQ',
                                   category_column='group')
study_test.run_model(draws)
study_test.plot_results()
```
